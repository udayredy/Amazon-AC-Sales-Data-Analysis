### Converting categorical data into numeric

#### I have found a "Get" value in the ratings column which is representing '0' as it has no ratings till now.

```python
Ac_data[Ac_data['ratings']=='Get']
```

```python
Ac_data[Ac_data['ratings']=='Get']["ratings"].shape
```

```python
Ac_data["ratings"].replace('Get','0',inplace=True)
```

```python
Ac_data[Ac_data['ratings']=='Get']["ratings"].shape
```

## Removing special Characters from the numerical data

```python
Ac_data[["discountprice","actualprice"]]=Ac_data[["discount_price","actual_price"]].replace(["₹",","],"",regex=True)
Ac_data[["Ratings","No_of_ratings"]]=Ac_data[["ratings","no_of_ratings"]].replace(["₹",","],"",regex=True)
```

```python
Ac_data
```

```python
Ac_data.drop(["image","link","discount_price","actual_price","ratings","no_of_ratings"],axis=1,inplace=True)
```

### Replacing Null values with median values

```python
Ac_data['No_of_ratings'].replace(['Only 2 left in stock.','Only 1 left in stock.'],['0','1'],inplace=True)
```

```python
Ac_data["No_of_ratings"].unique()
```

```python
Ac_data["discountprice"].fillna(Ac_data["discountprice"].median(),inplace=True)
Ac_data["actualprice"].fillna(Ac_data["actualprice"].median(),inplace=True)
Ac_data["Ratings"].fillna(Ac_data["Ratings"].median(),inplace=True)
Ac_data["No_of_ratings"].fillna(Ac_data["No_of_ratings"].median(),inplace=True)
```

```python
Ac_data[["discountprice","actualprice","Ratings","No_of_ratings"]].isnull().sum()
```

### Data type conversion from categorical to numerical

```python
Ac_data["discountprice"]= pd.to_numeric(Ac_data["discountprice"])
Ac_data["actualprice"]= pd.to_numeric(Ac_data["actualprice"])
Ac_data["Ratings"]= pd.to_numeric(Ac_data["Ratings"])
Ac_data["No_of_ratings"]= pd.to_numeric(Ac_data["No_of_ratings"])
```
